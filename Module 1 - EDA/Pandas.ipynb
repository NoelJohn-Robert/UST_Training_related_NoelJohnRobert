# 7. Handling Missing data

```python
import numpy as np
```

```python
# we replace ? with nan
# we set inplace as True to modify existing variable itself
# use false if we are planning to store to another variable
raw_data.replace("?", np.nan, inplace=True)
```

```python
# returns number data missing from each column
raw_data.isnull().sum()
```

```python
# percent of records missing
raw_data.isnull().sum() / raw_data.shape[0] * 100
```

```python
# handling missing values
# getting rid of the first 2 columns
# axis=1 for column
raw_data.drop(['symboling', 'normalized-losses'], axis=1, inplace=True)

# below error due to me, already removed columns
```

```python
raw_data["price"].dtype
```

```python
raw_data["price"] = raw_data["price"].astype("float64")
```

```python
# copying content to another dataframe
vehicle_df = raw_data.copy()
```

```python
vehicle_df["price"].median()
```

```python
# fills NaN values (which were missing) with the given value(median here)
vehicle_df["price"] = vehicle_df["price"].fillna(vehicle_df["price"].median())
```

```python
# drop rows with missing values
vehicle_df.dropna(inplace=True)
```

```python
# we dropped rows with missing values
# dropping is not good, as we lose actual data
vehicle_df.isnull().sum()
```

# 8. Drop duplicate values

```python
vehicle_df.duplicated()
```

```python
# no rows here mean no duplicate entries - use drop_duplicated() to handle duplicate rows
vehicle_df[vehicle_df.duplicated()]
```

```python
# adding a duplicate entry into the DataFrame

# select random sample form dataframe using frac, takes values in percentage
# parameter can also be n, denoting no. of samples
sample_df = vehicle_df.sample(frac=0.05)
vehicle_df1 = pd.concat([vehicle_df, sample_df], ignore_index=True)
```

```python
# these are the records we added as duplicate
vehicle_df1[vehicle_df1.duplicated()]
```

```python
# getting rid of all duplicates
vehicle_df1.drop_duplicates(inplace=True)
```

# 13. Merge

```python
# we are merging two dataframes|series
# inner herge by default
# we merge on a common column

df_series1 = pd.Series({"US": 2000, "UK": 1450, "India":134}, name="gdp_per_capita")
df_series2 = pd.Series({"US": 14000, "UK": 18000, "India":36000}, name="population")
```

```python
df1 = pd.DataFrame(df_series1).reset_index()
df1.columns = ["country", "gdp_per_capita"]
```

```python
df2 = pd.DataFrame(df_series2).reset_index()
df2.columns = ["country", "population"]
```

```python
pd.merge(df1, df2, on="country", how="inner")
```

```python
pd.merge(df1, df2, on="country", how="outer")
```
