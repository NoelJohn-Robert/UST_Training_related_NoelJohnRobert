# 2. Restaurants with online orders

```python
# zomato_df['online_order'].unique()  # from this, we get the unique values
# zomato_df['online_order'].value_counts()  # this gives values with count

online_order_yes = len(zomato_df[zomato_df['online_order'] == "Yes"]) / len(zomato_df) * 100
print(f"Online orders yes: {online_order_yes}")
print(f"Online orders no: {100-online_order_yes}")
print()

plt.pie(zomato_df['online_order'].value_counts(), autopct="%0.02f%%", labels=["Yes", "No"], colors=["Red", "Pink"])
plt.legend()
plt.show()

# better alternate for percentage
# zomato_df['online_order'].value_counts() / zomato_df.shape[0] * 100
# also, labels = zomato_df['online_order'].unique()
```

# 6. Distribution of Restaurant cost (for 2 people)

```python
# zomato_df['approx_cost(for two people)'].value_counts()

zomato_df['approx_cost(for two people)'] = zomato_df['approx_cost(for two people)'].str.replace(",", '').astype('float64')
```

```python
plt.figure(figsize=(30, 10))

plt.subplot(1, 2, 1)
sns.distplot(zomato_df['approx_cost(for two people)'], color='blue')

plt.subplot(1, 2, 2)
sns.boxplot(zomato_df['approx_cost(for two people)'], orient="h")

plt.show()
```

```python
# insight: approx cost for 2 person is positively skewed
```

```python
# finding median
median = zomato_df['approx_cost(for two people)'].median()
print(f"Median: {median}")

# finding 1st quantile
q1 = zomato_df['approx_cost(for two people)'].quantile(0.25)
print(f"First quartile value: {q1}")

# finding 3rd quantile
q3 = zomato_df['approx_cost(for two people)'].quantile(0.75)
print(f"Third quartile value: {q3}")

iqr = q3-q1
print(f"InterQuartile range: {iqr}")
```

```python
# Approx cost of 2 people range from 300 to 650 INR. 
```

```python
from geopy.geocoders import Nominatim
locations = pd.DataFrame({'Name':zomato_df['location'].unique()})
locations['Name'] = locations['Name'].apply(lambda x: "Bangalore "+str(x))
# locations
geolocator = Nominatim(user_agent="app")

lat_long = []
count = 0
for location in locations:
    location = geolocator.geocode(location)
    if location is None:
        lat_long.append(np.nan)
    else:
        geo = (location.latitude, location.longitude)
        lat_long.append(geo)
    count += 1
    if count == 10:
        break
print(lat_long)
```
